```python
n=11
k=0
fp_recall = []
fp_precision=  []
fn_recall = []
fn_precision=  []
err_type=0
score_type=0
for i in range(2,44):
    if i<2:
        continue
    temp_fp_recall=[]
    temp_fp_precision=[]
    temp_fn_recall=[]
    temp_fn_precision=[]
    if i==2:
        for j in range(5,16):
            temp_fp_recall.append(sheet.cell_value(i, j))
            temp_fn_recall.append(sheet.cell_value(i, j))
        fp_recall.append(temp_fp_recall)
        fn_recall.append(temp_fn_recall)
        continue
    if i==3:
        for j in range(5,16):
            temp_fp_precision.append(sheet.cell_value(i, j))
            temp_fn_precision.append(sheet.cell_value(i, j))
        fp_precision.append(temp_fp_precision)
        fn_precision.append(temp_fn_precision)
        continue
    else:
        if err_type%2==0:
            if i%2==0:
                for j in range(5,16):
                    temp_fp_recall.append(sheet.cell_value(i,j))
                fp_recall.append(temp_fp_recall)
            else:
                for j in range(5,16):
                    temp_fp_precision.append(sheet.cell_value(i,j))
                fp_precision.append(temp_fp_precision)
                err_type=err_type+1
        else:
            if i%2==0:
                for j in range(5,16):
                    temp_fn_recall.append(sheet.cell_value(i,j))
                fn_recall.append(temp_fn_recall)
            else:
                for j in range(5,16):
                    temp_fn_precision.append(sheet.cell_value(i,j))
                fn_precision.append(temp_fn_precision)
                err_type=err_type+1
print(fp_recall)
print(fn_recall)
print(fp_precision)
print(fn_precision)
```
